```python
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  #MacOS自带字体
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
import seaborn as sns
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df = pd.read_csv('data/tips.csv')
```

```python
df.head()
```

```python
df_smokerNo = df.groupby('smoker').get_group("No")
df_smokerYes = df.groupby('smoker').get_group("Yes")
```

```python
df_smokerYes.head()
```

```python
df_smokerNo['tip'].max()
```

```python
df_smokerYes['tip'].max()
```

```python
with plt.style.context('_mpl-gallery'):
    fig, axes = plt.subplots(1,2,figsize=(12,6))
    sns.kdeplot(df_smokerNo['tip'],label='No', ax=axes[0], fill=True, color='green')
    axes[0].legend()
    axes[0].set_title('不吸烟者小费分布')
    axes[0].set_ylabel('概率密度')
    sns.kdeplot(df_smokerYes['tip'],label='Yes', ax=axes[1], fill=True, color='red')
    axes[1].legend()
    axes[1].set_title('吸烟者小费分布')
    axes[1].set_ylabel('概率密度')
    plt.show()
```
