```python
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  #MacOS自带字体
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
import seaborn as sns
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import datetime
```

```python
df = pd.read_csv('data/dowjones.csv')
```

```python
t_range = pd.date_range('1914-12-01', '1968-12-01', freq='MS')
t_range.__len__()
s_test = pd.Series(np.random.randn(t_range.__len__()), index=t_range)
print(s_test.head(10))
print(s_test.resample('3MS').mean().head())
```

```python
s = Series(df['Price'].values,index=pd.to_datetime(df['Date'].values))
```

```python
with plt.style.context('Solarize_Light2'):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    s.plot(ax=axes[0])
    axes[0].set_title('按月')
    s.groupby(s.index.year).mean().plot(ax=axes[1])
    axes[1].set_title('按年')
    plt.tight_layout()
    plt.show()
```
